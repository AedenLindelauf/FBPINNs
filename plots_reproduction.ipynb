```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# load csv with the tensorboard data
df = pd.read_csv(r"C:\Users\35989\Desktop\University\MSc Applied Mathematics\SciML\Final Project\multi_level_training_results.csv")

steps = df.iloc[:,1]
lvl_124 = df.iloc[:,2]
lvl_4 = df.iloc[:,3]
lvl_12 = df.iloc[:,4]
lvl_2 = df.iloc[:,5]

plt.style.use("default")

fig, ax = plt.subplots(figsize=(6,5), facecolor="white")
ax.set_facecolor("white")

colors = {
    "124": "#1f77b4",   
    "4": "#7f7f7f",     
    "12": "#4fa3d1",    
    "2": "#bdbdbd"    
}

ax.semilogy(steps, lvl_124, color=colors["124"], linewidth=3, label="FBPINN [1,2,4]")
ax.semilogy(steps, lvl_4, color=colors["4"], linewidth=3, label="FBPINN [4]")
ax.semilogy(steps, lvl_12, color=colors["12"], linewidth=3, label="FBPINN [1,2]")
ax.semilogy(steps, lvl_2, color=colors["2"], linewidth=3, label="FBPINN [2]")

ax.set_xlabel("Training iterations", fontsize=13)
ax.set_ylabel("Normalised L1 test loss", fontsize=13)

ax.grid(True, which="major", linestyle="--", linewidth=0.6, alpha=0.5)

legend = ax.legend(frameon=True, fontsize=11)
legend.get_frame().set_edgecolor("black")
legend.get_frame().set_linewidth(0.8)

ax.tick_params(axis='both', labelsize=11)

plt.tight_layout()
plt.show()
```
