```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# load csv with the tensorboard data
df = pd.read_csv(r"C:\Users\35989\Desktop\University\MSc Applied Mathematics\SciML\Final Project\multi_level_training_results.csv")

steps = df.iloc[:,1]
lvl_124 = df.iloc[:,2]
lvl_4 = df.iloc[:,3]
lvl_12 = df.iloc[:,4]
lvl_2 = df.iloc[:,5]

plt.style.use("default")

fig, ax = plt.subplots(figsize=(6,5), facecolor="white")
ax.set_facecolor("white")

colors = {
    "124": "#1f77b4",   
    "4": "#7f7f7f",     
    "12": "#4fa3d1",    
    "2": "#bdbdbd"    
}

ax.semilogy(steps, lvl_124, color=colors["124"], linewidth=3, label="FBPINN [1,2,4]")
ax.semilogy(steps, lvl_4, color=colors["4"], linewidth=3, label="FBPINN [4]")
ax.semilogy(steps, lvl_12, color=colors["12"], linewidth=3, label="FBPINN [1,2]")
ax.semilogy(steps, lvl_2, color=colors["2"], linewidth=3, label="FBPINN [2]")

ax.set_xlabel("Training iterations", fontsize=13)
ax.set_ylabel("Normalised L1 test loss", fontsize=13)

ax.grid(True, which="major", linestyle="--", linewidth=0.6, alpha=0.5)

legend = ax.legend(frameon=True, fontsize=11)
legend.get_frame().set_edgecolor("black")
legend.get_frame().set_linewidth(0.8)

ax.tick_params(axis='both', labelsize=11)

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# load csv
df = pd.read_csv(r"C:\Users\35989\Desktop\University\MSc Applied Mathematics\SciML\Final Project\2D_width_training_3x64.csv")

steps = df["Step"]

w2 = df["width2"]
w4 = df["width4"]
w8 = df["width8"]
w16 = df["width16"]
pinn = df["3x64"]

plt.style.use("default")

fig, ax = plt.subplots(figsize=(6,5), facecolor="white")
ax.set_facecolor("white")

colors = {
    "pinn": "#7f7f7f",
    "w2": "#fdd0c7",
    "w4": "#fcae91",
    "w8": "#fb6a4a",
    "w16": "#cb181d"
}

ax.plot(steps, pinn, color=colors["pinn"], linewidth=3,
        label="PINN 3 layers 64 hidden units")

ax.plot(steps, w2, color=colors["w2"], linewidth=3,
        label="FBPINN 1 layer 2 hidden units")

ax.plot(steps, w4, color=colors["w4"], linewidth=3,
        label="FBPINN 1 layer 4 hidden units")

ax.plot(steps, w8, color=colors["w8"], linewidth=3,
        label="FBPINN 1 layer 8 hidden units")

ax.plot(steps, w16, color=colors["w16"], linewidth=3,
        label="FBPINN 1 layer 16 hidden units")

ax.set_xlabel("Training step", fontsize=13)
ax.set_ylabel("Training rate (steps/s)", fontsize=13)

ax.set_title("Training rate comparison", fontsize=14)

ax.grid(True, linestyle="--", linewidth=0.6, alpha=0.5)

legend = ax.legend(frameon=True, fontsize=11)
legend.get_frame().set_edgecolor("black")
legend.get_frame().set_linewidth(0.8)

ax.tick_params(axis='both', labelsize=11)

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# load csv with the tensorboard data
df = pd.read_csv(r"C:\Users\35989\Desktop\University\MSc Applied Mathematics\SciML\Final Project\results_1d_poisson.csv")

steps = df.iloc[:,1]
lvl_16 = df.iloc[:,2]
pinn_3x64 = df.iloc[:,3]
lvl_2_4_8_16 = df.iloc[:,4]
lvl_2 = df.iloc[:,5]
lvl_1_2 = df.iloc[:,6]
lvl_4 = df.iloc[:,7]
lvl_1_2_4 = df.iloc[:,8]

plt.style.use("default")

fig, ax = plt.subplots(figsize=(7,5), facecolor="white")
ax.set_facecolor("white")

ax.semilogy(steps, pinn_3x64, linewidth=3, label="PINN 3x64")
ax.semilogy(steps, lvl_16, linewidth=3, label="FBPINN [16]")
ax.semilogy(steps, lvl_4, linewidth=3, label="FBPINN [4]")
ax.semilogy(steps, lvl_2, linewidth=3, label="FBPINN [2]")
ax.semilogy(steps, lvl_2_4_8_16, linewidth=3, label="FBPINN [2,4,8,16]")
ax.semilogy(steps, lvl_1_2, linewidth=3, label="FBPINN [1,2]")
ax.semilogy(steps, lvl_1_2_4, linewidth=3, label="FBPINN [1,2,4]")

ax.set_xlabel("Training iterations", fontsize=13)
ax.set_ylabel("L1 test loss", fontsize=13)

ax.grid(True, which="major", linestyle="--", linewidth=0.6, alpha=0.5)

legend = ax.legend(frameon=True, fontsize=11)
legend.get_frame().set_edgecolor("black")
legend.get_frame().set_linewidth(0.8)

ax.tick_params(axis='both', labelsize=11)

plt.tight_layout()
plt.show()
```
